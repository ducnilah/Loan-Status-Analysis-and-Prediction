Part 1: Transform and prepare a csv file for training and testing predictive models.

```python
import pandas as pd
import zipfile
import kaggle
```

```python
!kaggle datasets download -d bhavikjikadara/loan-status-prediction
```

```python
zip_file = 'loan-status-prediction.zip'
with zipfile.ZipFile(zip_file, 'r') as z:
    z.extractall()
```

```python
dataset = pd.read_csv('loan_data.csv')
```

```python
dataset.head(10)
```

```python
# Convert categorical columns to numeric (male to 1, female to 0 for example)

mapping_gender = {'Female': 0, 'Male': 1}
mapping_married = {'No': 0, 'Yes': 1}
mapping_education = {'Not Graduate': 0, 'Graduate': 1}
mapping_self_employed = {'No': 0, 'Yes': 1}
mapping_property_area = {'Rural': 0, 'Semiurban': 1, 'Urban': 2}
mapping_loan_status = {'N': 0, 'Y': 1}

dataset['Gender'] = dataset['Gender'].map(mapping_gender)
dataset['Married'] = dataset['Married'].map(mapping_married)
dataset['Education'] = dataset['Education'].map(mapping_education)
dataset['Self_Employed'] = dataset['Self_Employed'].map(mapping_self_employed)
dataset['Property_Area'] = dataset['Property_Area'].map(mapping_property_area)
dataset['Loan_Status'] = dataset['Loan_Status'].map(mapping_loan_status)

# In this step i could go with LabelEncoder, but I prefer to use mapping for better control and the number of columns needed to change is not that big.
```

```python
dataset.head(10)
```

```python
dataset.describe()
```

```python
dataset.dtypes
```

```python
# Deal with missing values by using KNN Imputer

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=4)
df = imputer.fit_transform(dataset.select_dtypes(include=['float64', 'int64']))
```

```python
df = pd.DataFrame(df, columns=dataset.select_dtypes(include=['float64', 'int64']).columns)
```

```python
df.head(50)
```

```python
# Save the dataset to a csv file for training and testing predcitive models.

df.to_csv('loan_data_cleaned.csv', index=False)
```

Part 2: Prepare .xlsx file for futher analysis and reporting

```python
# Concat the cleaned dataset with LoanID
loan_id = dataset['Loan_ID']
cleaned_dataset = pd.concat([loan_id, df], axis=1)
```

```python
cleaned_dataset.head(10)
```

```python
# Convert the cleaned dataset to original format

mapping_reversed_gender = {1: 'Male', 0: 'Female'}
mapping_reversed_married = {1: 'Yes', 0: 'No'}
mapping_reversed_education = {1: 'Graduate', 0: 'Not Graduate'}
mapping_reversed_self_employed = {1: 'Yes', 0: 'No'}
mapping_reversed_property_area = {0: 'Rural', 1: 'Semiurban', 2: 'Urban'}
mapping_reversed_loan_status = {1: 'Y', 0: 'N'}

cleaned_dataset['Gender'] = cleaned_dataset['Gender'].map(mapping_reversed_gender)
cleaned_dataset['Married'] = cleaned_dataset['Married'].map(mapping_reversed_married)
cleaned_dataset['Education'] = cleaned_dataset['Education'].map(mapping_reversed_education)
cleaned_dataset['Self_Employed'] = cleaned_dataset['Self_Employed'].map(mapping_reversed_self_employed)
cleaned_dataset['Property_Area'] = cleaned_dataset['Property_Area'].map(mapping_reversed_property_area)
cleaned_dataset['Loan_Status'] = cleaned_dataset['Loan_Status'].map(mapping_reversed_loan_status)
```
